```python
df_holiday = pd.read_csv('US Holiday Dates (2004-2021).csv')
```

```python
df_holiday.head()
```

```python
df_holiday['Date'] = pd.to_datetime(df_holiday['Date'], format='%Y-%m-%d')
```

```python
def contains_holiday(start_date, end_date, holidays):
    mask = (holidays >= start_date) & (holidays <= end_date)
    holiday_count = mask.sum()
    
    if holiday_count > 0:
        return "Yes"
    else:
        return "No"
```

```python
df['Contains Holiday'] = df.apply(lambda row: contains_holiday(row['From Date'], row['To Date'], df_holiday['Date']), axis=1)
```

```python
df_gas_price = pd.read_csv('Gasoline_Prices.csv')
```

```python
df_gas_price.describe()
```

```python
df_gas_price['Date'] = pd.to_datetime(df_gas_price['Date'], format='%m/%d/%Y')
```

```python
def get_gas_price(start_date, end_date, gas_prices):
    mask = (gas_prices['Date'] >= start_date) & (gas_prices['Date'] <= end_date)
    
    if mask.any():
        return gas_prices.loc[mask, 'New York State Average ($/gal)'].iloc[0]
    else:
        return None
```

```python
df['Gas Price'] = df.apply(lambda row: get_gas_price(row['From Date'], row['To Date'], df_gas_price), axis=1)
```

```python
df.head()
```

```python
columns_to_remove = ['Full Fare', 'Senior Citizen/Disabled', '7 Day ADA Farecard Access System Unlimited',
                     '30 Day ADA Farecard Access/Reduced Fare Media Unlimited', 'Joint Rail Road Ticket',
                     '7 Day Unlimited', '30 Day Unlimited', '14 Day Reduced Fare Media Unlimited', '1 Day Unlimited',
                     '14 Day Unlimited', '7 Day Express Bus Pass', 'TransitCheck', 'Long Island Bus Special Senior',
                     'Reduced Fare 2-Trip', 'Rail Road Unlimited No Trade', 'TransitCheck Annual',
                     'Mail and Ride EasyPay Express', 'Mail and Ride EasyPay Unlimited', 'PATH 2-Trip',
                     'AirTrain Full Fare', 'AirTrain 30 Day Unlimited', 'AirTrain 10-Trip', 'AirTrain Monthly',
                     'Student', 'NICE 2-Trip', 'CUNY 120 Day', 'CUNY 60 Day', 'Fair Fares Pay-Per-Ride',
                     'Fair Fares 7 Day Unlimited', 'Fair Fares 30 Day Unlimited']

df_final = df.drop(columns=columns_to_remove)
```

```python
for index, row in df_final.iterrows():
    from_date = row['From Date']
    to_date = row['To Date']
    
    matching_rows = weather_df_final[(weather_df_final['DATE'] >= from_date) & (weather_df_final['DATE'] <= to_date)]
    
    if not matching_rows.empty:
        total_snowfall = matching_rows['SNOW'].sum()
        total_rainfall = matching_rows['PRCP'].sum()
        
        df_final.at[index, 'Total Snowfall'] = total_snowfall
        df_final.at[index, 'Total Rainfall'] = total_rainfall
```
