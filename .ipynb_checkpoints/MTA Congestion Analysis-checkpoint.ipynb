```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create datafame out of the csv file.
df = pd.read_csv('final_dataframe_MTA_congestion.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
#Create histograms for all of the numeric features to get a sense of the disribution
#Added the KDE to get the estimated probability desnity function

numeric_features = ['Total Fares for the Week', 'Gas Price', 'Total Snowfall', 'Total Rainfall']

for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=df, x=feature, kde=True)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

When looking at the distributions of the data there are a few things to notice, almost all of the data is right skewed. This makes sense for the weather events as there are normally not extreme but does not explain why some weeks there are extreme values for the total fares. The distribution of the fares could be due to a time series. 

-------------------------------------------------------------------------------------------------------------------

```python
sns.pairplot(data=df, vars=numeric_features)
plt.title("Pairplot of Numeric Features", y=-0.35)
plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df[numeric_features].corr()
print("Pearson Correlation Coefficients:")
print(correlation_matrix)
```

The scatter plots dont really show strong correlations between the features and target, there is somewhat of a pattern in the scatterplots involving total fares, gas price, and the total rainfall, but no strong relationships. 

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Heatmap")
plt.show()
```

Looking at the linear correlations there does not seem to be any strong relationships between the features and target, there is a weak correlation between the total fares gas prices and total snowfall, and interestingly there is a negative correlation between the total fares and the rainfall. 

```python
#Boxplot to show if there is any relationship between holidays and total fares
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Contains Holiday', y='Total Fares for the Week')
plt.yscale('log')
plt.title("Total Fares by Holiday Presence")
plt.show()
```

```python
summary = df.groupby('Contains Holiday')['Total Fares for the Week'].describe().round(2)
print("Five-Number Summary:")
print(summary[['min', '25%', '50%', '75%', 'max']])
```

```python
means = df.groupby('Contains Holiday')['Total Fares for the Week'].mean().round(2)
print("Average Total Fares:")
print(means)
```
